```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
H = np.loadtxt('topography_180x360_grid.txt')
```

```python
plt.imshow(H)
```

```python
la,lo = np.shape(H)
print(la,lo)
```

```python
lat0 = 90
lon0 = 0
R = 6371
deg2rad = (np.pi/180)

total_area = 0
ocean_area = 0
ocean_volume = 0

latdist = R*1*deg2rad

for i in range(la):
    for j in range(lo):
        lat = lat0 - i
        lon = lon0 + j
        elevation = H[i,j]/1000.0
        
        londist = 2*np.pi*R*np.cos(lat*deg2rad)/lo
        
        d_area = londist*latdist
        total_area += d_area
        
        if elevation < 0:
            ocean_area += d_area
            ocean_volume += abs(d_area*elevation)
        
```

```python
true_area = 4*np.pi*R**2
```

```python
#Accuracy
total_area/true_area
```

```python
#ocean area
print(ocean_area)
```

```python
#fraction
ocean_area/total_area
```

```python
#volume of ocean
print('ocean volume: %.4g km^3' %ocean_volume)
```
